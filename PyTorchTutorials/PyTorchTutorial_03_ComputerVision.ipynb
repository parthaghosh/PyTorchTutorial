```python
# Utiltiy Functions 

# Version Checking 
def VersionCheck(): 
    # Check versions 
    print(f"torch.version: {torch.__version__}")
    print(f"torchvision.version: {torchvision.__version__}")
    pass 

def GetDeviceName():
    if (torch.cuda.is_available()): 
        device_name = "cuda"
    else:
        device_name = "cpu"
    return(device_name)

# Import library from a URL 
def ImportFromURL(fileName, urlLink=None):
    if (Path(fileName).is_file()): 
        print(f"File ({fileName}) already exists. Ignoring downloading. ")
    else:
        print(f"Downloading File ({fileName}) ... ")
        request = requests.get(urlLink)
        with open(fileName,"wb") as f:
            f.write(request.content)
    pass 

# Main ML Code

# Load Dataset 
def CrtBatchesFromDataset_Fashion(dstRootDir, batchSz):
    # Load Datasets from source
    dataset_trng = datasets.FashionMNIST(root=dstRootDir, train=True,  
                                         download=True, transform=ToTensor(), 
                                         target_transform=None)
    dataset_test = datasets.FashionMNIST(root=dstRootDir, train=False, 
                                         download=True, transform=ToTensor()) 
    # Create batches from Datasets
    batches_trng = DataLoader(dataset_trng, batch_size=batchSz, shuffle=True)
    batches_test = DataLoader(dataset_test, batch_size=batchSz, shuffle=False) # No need for shuffling test
    
    return ({'data_train': dataset_trng,'data_test': dataset_test,
             'batches_train': batches_trng,'batches_test': batches_test})
 
#
# Experiment with ML models. 
# 

#
# Define ML Linear model: V0 
#
class VisionModel_Linear_V0(nn.Module): 
    def __init__(self, inp_dim:int, hidden_dim:int, out_dim: int) -> None: 
        super().__init__()
        self._layer_stack = nn.Sequential(
            nn.Flatten(), # Required to create a single vector from 2D tensor
            nn.Linear(in_features=inp_dim, out_features=hidden_dim),
            nn.Linear(in_features=hidden_dim, out_features=out_dim)
            )
    pass 

    def PredictLabel(self,X): 
        out_tensor = self.forward(X)
        label = torch.softmax(out_tensor.squeeze(dim=0), dim=0).argmax(dim=0).item()
        return(label)

    def forward(self, X:torch.Tensor) -> None : 
        return (self._layer_stack(X))
    
    pass 


# Define ML ReLU model: V1 
class VisionModel_ReLU_V1(nn.Module): 
    def __init__(self, inp_dim:int, hidden_dim:int, out_dim: int) -> None: 
        super().__init__()
        self._layer_stack = nn.Sequential(
            nn.Flatten(), # Required to create a single vector from 2D tensor
            nn.Linear(in_features=inp_dim, out_features=hidden_dim), nn.ReLU(), 
            nn.Linear(in_features=hidden_dim, out_features=out_dim), nn.ReLU()
            )
    pass 

    def PredictLabel(self,X): 
        out_tensor = self.forward(X)
        label = torch.softmax(out_tensor.squeeze(dim=0), dim=0).argmax(dim=0).item()
        return(label)

    def forward(self, X:torch.Tensor) -> None : 
        return (self._layer_stack(X))
    
    pass 

# Define ML CNN model: V2
class VisionModel_CNN_V2(nn.Module): 
    def __init__(self, in_channels:int, hidden_dim:int, out_classes: int) -> None: 
        super().__init__()
        
        #
        # Input: [1,in_channels,64,64] => Output: [1,hidden_dim,31,31]
        # Formula: Output Volume = [(W+2P-K)/S]+1
        #
        self._layer1 = nn.Sequential( 
                            nn.Conv2d(in_channels=in_channels,  
                                 out_channels=hidden_dim, # 10-channels as output
                                 kernel_size=5, 
                                 stride=2, padding=1), 
                            nn.ReLU()
                    )
        
        # Input: [1,hidden_dim,31,31] => Output: [1,hidden_dim,8,8]
        self._layer2 = nn.Sequential( 
                            nn.Conv2d(in_channels=hidden_dim,    
                                 out_channels=hidden_dim,
                                 kernel_size=3, stride=2, padding=1), 
                            nn.ReLU(),
                            nn.MaxPool2d(kernel_size=2)  
                        )
        
        self._layer3  = nn.Sequential( # Input: [1,10,8,8] => Output: [1,10,10]
                            # Flatten(): Output becomes 640 = 10*8*8 for each channel => (1,640)
                            nn.Flatten(start_dim=1,end_dim=-1), 
                            nn.Linear(in_features=450, out_features=hidden_dim),
                            nn.ReLU(),
                            nn.Linear(in_features=hidden_dim, out_features=out_classes)
                      )
    pass 

    def PredictLabel(self,X): 
        out_tensor = self.forward(X)
        label = torch.softmax(out_tensor.squeeze(dim=0), dim=0).argmax(dim=0).item()
        return(label)

    def forward(self, X:torch.Tensor) -> None : 
        # Run by cascaded layers 
        x = self._layer1(X)
        x = self._layer2(x)
        x = self._layer3(x)
        return(x)
    pass 

class VisionModel_CNN_V2_AutoParameters(nn.Module): 
    def __init__(self, in_channels:int, in_img_width:int, hidden_dim:int, out_classes: int) -> None: 
        super().__init__()
          
        #
        # Input: [1,in_channels,64,64] => Output: [1,hidden_dim,31,31]
        #
        l1_in_channels = in_channels; l1_img_width =in_img_width;l1_hid_dims = hidden_dim
        l1_kernel_sz = 5; l1_stride_sz=2; l1_padding_sz=1
        l1_out_img_dim = int((l1_img_width+2*l1_padding_sz-l1_kernel_sz)/l1_stride_sz)+1
        #
        # Formula: Output Volume = [int((W+2P-K)/S)+1]
        #
        # Image Width 
        l1_out_shape = (l1_in_channels, hidden_dim, l1_out_img_dim, l1_out_img_dim)
        
        self._layer1 = nn.Sequential( 
                            nn.Conv2d(in_channels=l1_out_shape[0],  
                                 out_channels=l1_out_shape[1], # 10-channels as output
                                 kernel_size=l1_kernel_sz,
                                 stride=l1_stride_sz, padding=l1_padding_sz), 
                            nn.ReLU()
                    )
        
        # Input: [1,hidden_dim,31,31] => Output: [1,hidden_dim,8,8]
        l2_img_width = l1_out_shape[2]
        l2_kernel_sz = 3; l2_stride_sz=2; l2_padding_sz=1
        l2_in_channels = l1_out_shape[1]; 
        l2_hid_dims = hidden_dim
        l2_out_conv_img_dim = int((l2_img_width+2*l2_padding_sz-l2_kernel_sz)/l2_stride_sz)+1
        l2_maxpool_kernel_sz=2
        l2_out_maxpool_out_img_dim = int(l2_out_conv_img_dim/l2_maxpool_kernel_sz)
        l2_out_shape = (l2_in_channels, l2_hid_dims, l2_out_maxpool_out_img_dim, l2_out_maxpool_out_img_dim)
        self._layer2 = nn.Sequential( 
                            nn.Conv2d(in_channels=l2_out_shape[0],    
                                 out_channels=l2_out_shape[1],
                                 kernel_size=l2_kernel_sz, stride=l2_stride_sz, 
                                      padding=l2_padding_sz), 
                            nn.ReLU(),
                            nn.MaxPool2d(kernel_size=l2_maxpool_kernel_sz)  
                        )
        
        l3_linear_inp_width = l2_out_shape[1]*l2_out_shape[2]*l2_out_shape[3]
        l3_hid_dims = l2_out_shape[1]
        l3_out_dim = out_classes
        
        self._layer3  = nn.Sequential( # Input: [1,10,8,8] => Output: [1,10,10]
                            # Flatten(): Output becomes 640 = 10*8*8 for each channel => (1,640)
                            nn.Flatten(start_dim=1,end_dim=-1), 
                            nn.Linear(in_features=l3_linear_inp_width, out_features=l3_hid_dims),
                            nn.ReLU(),
                            nn.Linear(in_features=l3_hid_dims, out_features=out_classes)
                      )
    pass 

    def PredictLabel(self,X): 
        out_tensor = self.forward(X)
        label = torch.softmax(out_tensor.squeeze(dim=0), dim=0).argmax(dim=0).item()
        return(label)

    def forward(self, X:torch.Tensor) -> None : 
        # Run by cascaded layers 
        x = self._layer1(X)
        x = self._layer2(x)
        x = self._layer3(x)
        return(x)
    pass 

#
# Quick Check: Custom Loss Function 
# GUIDE: https://saturncloud.io/blog/custom-loss-function-in-pytorch-a-comprehensive-guide/
#
class MyLossFn (nn.Module):
    #_real_loss_fn = None
    def __init__(self): 
        super(MyLossFn,self).__init__() 
        self._real_loss_fn = torch.nn.CrossEntropyLoss()
        
    def forward(self, predicted, target):
        # loss = torch.mean((predicted - target) ** 2)
#         print("1. MyLossFn called *************\n")
        loss = self._real_loss_fn(predicted, target)
#         print("2. MyLossFn called *************\n")
        return loss
    
    # loss.backward() is a standard method within nn.Module()
    
pass  # End of MyLossFn 

# Loading and Saving models 
def SaveModel (modelObj:nn.Module, modelPath:Path, modelFname:str) -> None:
    MODEL_PATH = Path(modelPath)
    MODEL_SAVE_PATH = MODEL_PATH/modelFname
    # Now save the model
    torch.save(obj=modelObj.state_dict(), f=MODEL_SAVE_PATH)    
    pass 

def LoadModel (modelObj:nn.Module, modelSavePath:Path, deviceName:str)->None:
    # Now load the model
    modelObj.load_state_dict(torch.load(f=modelSavePath))
    return(modelObj).to(deviceName)


#
# Training Loop ....
# 
class TrainModel(object):
    _model_obj = None
    _default_learn_rate = 0.1
    
    # Constructor
    def __init__(self, modelObj:torch.nn, numClasses:int, 
                 device:torch.device, lossFn: torch.nn.Module = None, 
                 learnRate:float=None,optimizerFn: torch.optim.Optimizer = None 
                 ) -> None: 
        super().__init__() 
        self._device = device
 
        self._model_obj = modelObj.to(self._device) 
        self._model_name =modelObj.__class__.__name__     

        self._num_classes = numClasses
        
        if (lossFn==None): 
            self._loss_fn = torch.nn.CrossEntropyLoss()
        else:
            self._loss_fn = lossFn

        
        if (learnRate==None): 
            self._lr = self._default_learn_rate
        else:
            self._lr = learnRate
            
        if (optimizerFn==None): 
            self._optimizer = torch.optim.SGD(params=self._model_obj.parameters(), 
                                             lr=self._lr)
        else:
            self._optimizer = optimizerFn
        pass 
    
    # Training Loop 
    def train_step_xepochs(self, trainDataLoader:torch.utils.data.DataLoader, numEpochs:int) -> None: 
        # Define functions, variables etc
        loss_fn = self._loss_fn
        optimizer = self._optimizer
        device_name = self._device  
        debug_flag = False
                      
        # Outer training loop across epochs 
        total_examples = len(trainDataLoader.dataset)
        num_classes = self._num_classes
        num_epochs = numEpochs
        
        # Time End: End of training task 
        time_start = timer() 
       
        # Start EPOCHs 
        self._model_obj.train()
        for epoch in tqdm(range(num_epochs)): 
            train_loss, train_acc = 0, 0
            for batch, (X,y_targets) in enumerate(trainDataLoader):
                X, y_targets = X.to(device_name), y_targets.to(device_name)   
                
                # forward pass. device is already determined at __init__(). 
                y_preds = self._model_obj(X)                 
                
                train_batch_loss = loss_fn(y_preds, y_targets) # batch loss 
                train_acc += torchmetrics.functional.accuracy(y_preds, y_targets, task="multiclass",num_classes=num_classes)
                train_loss += train_batch_loss # overall loss across batches 
                optimizer.zero_grad() 
                train_batch_loss.backward() # backpropagation
                optimizer.step() # Update params 
                if debug_flag==True: 
                    if (batch%1000==0): 
                        print(f"Batch#{batch}: Looked at {batch*len(X)}/{total_examples} samples..  ")
            # Average training loss across batches
            num_batches = len(trainDataLoader)
            train_loss /= num_batches
            train_acc /= num_batches
            model_name = self._model_name 
            
            print(f"Epoch#{epoch}:\tModel Name:{model_name}\tTRAIN\t[loss:{train_loss:.3f}\tacc:{train_acc:.2f}]")   

        # Time End: End of training task 
        time_end = timer() 
        time_spent = time_end - time_start 
        
        print(f"train_step_xepochs(): Duration : {time_spent:.2f} seconds")

        pass  # train_step() 


    # Training Loop 
    def test_step(self, testDataLoader:torch.utils.data.DataLoader) -> None : 
        
        loss_fn = self._loss_fn
        optimizer = self._optimizer
        device_name = self._device 
        debug_flag = False
                      
        # Outer test loop             
        self._model_obj.eval()

        num_test_examples = len(testDataLoader)
        test_loss, test_acc =0,0

        with torch.inference_mode(): 
            for X,y_targets in testDataLoader:

                X, y_targets  = X.to(device_name), y_targets.to(device_name)
                y_preds = self._model_obj(X) # forward pass. device is already determined at __init__(). 
            
                test_loss += loss_fn(y_preds, y_targets)
                test_acc += torchmetrics.functional.accuracy(y_preds, y_targets, task="multiclass", num_classes=num_classes)

            # Average test_loss
            test_loss /= num_test_examples
            test_acc /= num_test_examples
                 
            print(f"TEST:\t[loss:{test_loss:.3f}\tacc:\t{test_acc:.2f}]")
            
        pass # class test_step


# Plot image samples 
# # Args: (dset_test, sample_indices, pred_labels)        
def PlotImageSamples(dataSetCorpus, sampleIndices, predLabels=None):
    # Total samples images 
    n_samples = len(sampleIndices)

    # Predictor class names
    class_names = dataSetCorpus.classes
    
    # Define subplot regions
    num_rows = int(np.round(np.sqrt(n_samples)))+1
    num_cols = num_rows
    
    # Start plotting 
    plt.figure(figsize=(17,17))
    for i in range(n_samples): 
        img_index = sampleIndices[i]
        img, true_label = dataSetCorpus[img_index]
        title_str = f"{class_names[true_label]}"
        if (predLabels!=None):
            pred_label = predLabels[i]
            if (pred_label!=true_label): 
                title_str = ("True:"+title_str + (f"=>Pred:{class_names[pred_label]}"))
        plt.subplot(num_rows,num_cols,i+1)
        plt.title(title_str)
        plt.axis(False)
        plt.imshow(img.squeeze(dim=0), cmap='gray')
    pass 

# 
# Plot confusion matrix using torchmetrics. 
#
def PlotConfMatrix(classList:List[str], predLabels:List[int], trueLabels:List[int])->None:
    # How many classes?
    class_list = classList
#     print(f"PlotConfMatrix: class_list\t{class_list}\n")
#     print(f"PlotConfMatrix: pred_labels\t{predLabels}\n")
#     print(f"PlotConfMatrix: true_labels\t{trueLabels}\n")
    num_classes = len(class_list)
    
    # Convert labels to tensors for Confusion Matrix 
    pred_label_tensors = torch.tensor(predLabels) 
    true_label_tensors = torch.tensor(trueLabels) 
    
    # Create confusion matrix with data (tensors).
    confusion_mat = ConfusionMatrix(num_classes=num_classes, task='multiclass')
    conf_mat_tensors = confusion_mat(preds=pred_label_tensors,target=true_label_tensors)
    
    # Plot the confusion matrix. 
    conf_mat_numpy = conf_mat_tensors.numpy()
    plot_confusion_matrix(conf_mat=conf_mat_numpy,class_names=class_list,figsize=(10, 7))
    plt.show()
    
    pass # End of PlotConfMatrix

#
# Main Routines 
#
dst_data_dir = "C:\\ParthaGhosh\\Technical\\MachineLearning\\DataSet\\Fashion_MNIST"

#
# Returns: {  'data_train': dataset_trng,    'data_test': dataset_test,
#             'batches_train': batches_trng, 'batches_test': batches_test}
BATCH_SZ = 32
response  = CrtBatchesFromDataset_Fashion(dst_data_dir,BATCH_SZ)
dset_trng = response['data_train']
dset_test = response['data_test'] 
batches_trng = response['batches_train']
batches_test = response['batches_test'] 

# How many classes under this problem? 
img_classes = dset_trng.classes

# # Plot pictures 
# # Sample few pictures and plot. 
# num_samples = 100 
# sample_indices = torch.
# PlotImageSamples(dset_trng, 100)

# Define Model.
# Compute image dimensions to determine size of input
img, label = dset_trng[0]

# Image Shape: Channels:Ht:Wdth

# Required for CNN models.
img_channels = img.shape[0]
img_ht = img.shape[1]
img_wdth = img.shape[2]

# Required for Linear models.
dim_tuple = img.squeeze(dim=0).shape
inp_dim = dim_tuple[0]*dim_tuple[1]

# Common to any models
out_dim = len(img_classes)
num_classes = len(dset_trng.classes)
hid_dim = 50 

# Deploy ML model

# Which device?
device_name = GetDeviceName()
in_img_width = img.shape[1]  

# model_vision = VisionModel_Linear_V0(inp_dim,hid_dim,out_dim)
# model_vision = VisionModel_ReLU_V1(inp_dim,hid_dim,out_dim)
# model_vision = VisionModel_CNN_V2(in_channels=img_channels, 
#                                   hidden_dim=hid_dim,
#                                   out_classes=num_classes)
model_vision = VisionModel_CNN_V2_AutoParameters(in_channels=img_channels, 
                                   in_img_width = in_img_width,   
                                  hidden_dim=hid_dim,
                                  out_classes=num_classes)

# TRAIN CYCLE 

# Instantiate and Train the model 
num_epochs, lr = 2, 0.075
# num_classes = len(dset_trng.classes)

#
# Signature of TrainModel() is as follows: 
# self, modelObj:torch.nn, numClasses:int, device:torch.device,
#                  lossFn: torch.nn.Module = None, 
#                  learnRate:float=None,
#                  optimizerFn: torch.optim.Optimizer = None                     
#

# A flag that asserts how the TrainModel() is to be instantiated. 
cfg_loss_optm_flag = True  

if (cfg_loss_optm_flag==True):

    # Various STANDARD optimizers
    optimizer = torch.optim.SGD(params=model_vision.parameters(), lr=lr)
    # optimizer = torch.optim.Adam(params=model_vision.parameters(), lr=lr)

    # Custom Loss Function Instantiation
    loss_fn = MyLossFn()
    # loss_fn = torch.nn.CrossEntropyLoss()
    
    # Call training loop 
    train_obj = TrainModel(model_vision, num_classes, device_name, 
                           lossFn=loss_fn,optimizerFn=optimizer )
#   train_obj = TrainModel(model_vision, num_classes, device_name, 
#                            lossFn=loss_fn,optimizerFn=optimizer )
else: 
    train_obj = TrainModel(model_vision, num_classes, device_name)

# trainDataLoader, testDataLoader
train_obj.train_step_xepochs(batches_trng, num_epochs)
train_obj.test_step(batches_test)

# PREDICTION/TEST CYCLE 

# Show sampled images and their predictions.
num_samples = 20
# Create test sample, true labels and predicted labels 
num_test_imgs = len(dset_test)
sample_indices = np.random.randint(low=0, high=num_test_imgs, size=num_samples)
pred_labels  = []
true_labels = []

# Set model to evaluation mode.  
model_vision.eval() 

# Start iterating predictions on list of vectors  
for sample_index in sample_indices: 
    img_test, true_label = dset_test[sample_index]
    true_labels.append(true_label)  # Even though not used immediately. 
    # Predict labels 
    X = img_test.unsqueeze(dim=0).to(device_name)
    pred_label = model_vision.PredictLabel(X)
    pred_labels.append(pred_label)  # Used in PlotImageSamples()

PlotImageSamples(dset_test, sample_indices, pred_labels)    
# If you don't want to predict, don't pass pred_labels as follow.   
# PlotImageSamples(dset_test, sample_indices)

# Plot confusion matrix using torchmetrics. 
# Signature: PlotConfMatrix(classList:List[str], predLabels:List[int], trueLabels:List[int])
PlotConfMatrix(img_classes, pred_labels, true_labels)


# Flag to enable testing of loading/savig model. 
Flag_model_load_save = False 

#
# Example: Model Saving and loading.
# NOTE: It has some problem due to read access to model file saved from
# this program. This can be overcome by logging in as enterprise user. 
#
if (Flag_model_load_save==True):
    # Create a model repository  
    model_path = "C:\\ParthaGhosh\\MyDocuments"
    model_f_name = "vision_model.pt"

    # Save the model 
    SaveModel(model_vision,model_path, model_f_name)

    # Load the model by instantiating first  
    model_vision = VisionModel_CNN_V2_AutoParameters(in_channels=img_channels, 
                                       in_img_width = in_img_width,   
                                      hidden_dim=hid_dim,
                                      out_classes=num_classes)

    MODEL_PATH = Path(model_path) ; MODEL_SAVE_PATH = MODEL_PATH/model_f_name
    device_name = GetDeviceName()

    # Load the weights into model instance as follows. 
    model_vision = LoadModel(model_vision,MODEL_PATH,device_name)
```
